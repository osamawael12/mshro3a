Create a pie chart to visualize the distribution of 'Sale Price' by 'Customer Gender'.

## Aggregate sales by gender

### Subtask:
Group the dataframe by 'Customer Gender' and calculate the sum of 'Sale Price' for each gender.


**Reasoning**:
Group the dataframe by 'Customer Gender' and calculate the sum of 'Sale Price' for each gender, then store the result in a new dataframe.



```python
sales_by_gender = df.groupby('Customer Gender')['Sale Price'].sum().reset_index()
display(sales_by_gender)
```

## Create the pie chart

### Subtask:
Generate a pie chart showing the proportion of sales for each gender.


**Reasoning**:
Generate a pie chart showing the proportion of sales for each gender using the pre-calculated `sales_by_gender` dataframe.



```python
plt.figure(figsize=(8, 8))
plt.pie(sales_by_gender['Sale Price'], labels=sales_by_gender['Customer Gender'], autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title('Distribution of Sale Price by Customer Gender')
plt.show()
```

## Summary:

### Data Analysis Key Findings

* The sales by gender distribution is split between female and male customers.
* The total sale price for female customers is slightly higher than for male customers.

* Further analysis could explore if there are specific product categories or purchase behaviors that contribute to the difference in sales between genders.


# Task
Create two charts: one showing the sum of 'Sale Price' by 'Delivery Type' and another showing the sum of 'Sale Price' by 'Product Category'.

## Aggregate sales by delivery type

### Subtask:
Group the dataframe by 'Delivery Type' and calculate the sum of 'Sale Price' for each delivery type.


**Reasoning**:
Group the dataframe by 'Delivery Type' and calculate the sum of 'Sale Price' for each delivery type, then store the result in a new dataframe and display it.



```python
sales_by_delivery_type = df.groupby('Delivery Type')['Sale Price'].sum().reset_index()
display(sales_by_delivery_type)
```

## Create chart for delivery type

### Subtask:
Generate a chart to visualize the sales distribution by delivery type.


**Reasoning**:
Generate a bar chart showing the total sale price for each delivery type using the pre-calculated `sales_by_delivery_type` dataframe.



```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Delivery Type', y='Sale Price', data=sales_by_delivery_type, palette='viridis')
plt.title('Total Sale Price by Delivery Type')
plt.xlabel('Delivery Type')
plt.ylabel('Total Sale Price')
plt.show()
```

## Aggregate sales by product category

### Subtask:
Group the dataframe by 'Product Category' and calculate the sum of 'Sale Price' for each product category.


**Reasoning**:
Group the dataframe by 'Product Category' and calculate the sum of 'Sale Price' for each product category, then store the result in a new dataframe and display it.



```python
sales_by_product_category = df.groupby('Product Category')['Sale Price'].sum().reset_index()
display(sales_by_product_category)
```

## Create chart for product category

### Subtask:
Generate a chart to visualize the sales distribution by product category.


**Reasoning**:
Generate a bar plot to visualize the sales distribution by product category using the pre-calculated `sales_by_product_category` dataframe.



```python
plt.figure(figsize=(12, 7))
sns.barplot(x='Product Category', y='Sale Price', data=sales_by_product_category, palette='viridis')
plt.title('Total Sale Price by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sale Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   Standard Delivery had the highest total sale price among the delivery types.
*   The total sale price for 'Shipped from Abroad' and 'Express' delivery types were significantly lower than that of 'Standard Delivery'.
*   The total sale price varied significantly across different product categories.

### Insights or Next Steps

*   Investigate the reasons behind the difference in sale prices across delivery types to potentially optimize delivery strategies.
*   Analyze the top-performing product categories further to understand what drives their success and identify opportunities for growth in other categories.


# Task
Create a line chart showing the sum of 'Order Quantity', 'Sale Price', and 'Profit' by 'OrderMonth'.

## Aggregate data by ordermonth

### Subtask:
Group the dataframe by 'OrderMonth' and calculate the sum of 'Order Quantity', the sum of 'Sale Price', and the sum of 'Profit' for each month.


**Reasoning**:
Group the dataframe by 'OrderMonth' and calculate the sum of 'Order Quantity', 'Sale Price', and 'Profit' for each month, then store the result in a new dataframe and display it.



```python
monthly_sales = df.groupby('OrderMonth')[['Order Quantity', 'Sale Price', 'Profit']].sum().reset_index()
display(monthly_sales.head())
```

## Create the line chart

### Subtask:
Generate a line chart showing the aggregated 'Order Quantity', 'Sale Price', and 'Profit' over the months.


**Reasoning**:
Generate a line chart showing the aggregated 'Order Quantity', 'Sale Price', and 'Profit' over the months using the pre-calculated `monthly_sales` dataframe.



```python
plt.figure(figsize=(12, 18))

# Plot Order Quantity
plt.subplot(3, 1, 1)
plt.plot(monthly_sales['OrderMonth'], monthly_sales['Order Quantity'], label='Order Quantity', color='blue')
plt.title('Monthly Trend in Order Quantity')
plt.xlabel('Month')
plt.ylabel('Order Quantity')
plt.legend()
plt.grid(True)

# Plot Sale Price
plt.subplot(3, 1, 2)
plt.plot(monthly_sales['OrderMonth'], monthly_sales['Sale Price'], label='Sale Price', color='orange')
plt.title('Monthly Trend in Sale Price')
plt.xlabel('Month')
plt.ylabel('Sale Price')
plt.legend()
plt.grid(True)

# Plot Profit
plt.subplot(3, 1, 3)
plt.plot(monthly_sales['OrderMonth'], monthly_sales['Profit'], label='Profit', color='green')
plt.title('Monthly Trend in Profit')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Summary:

### Data Analysis Key Findings

* The analysis successfully grouped the data by month and calculated the total 'Order Quantity', 'Sale Price', and 'Profit' for each month.
* A line chart was generated showing the trend of 'Order Quantity', 'Sale Price', and 'Profit' over time.

### Insights or Next Steps

* Analyze the line chart to identify any seasonal patterns or significant changes in order quantity, sales price, and profit over the months.
* Further investigate months with notable deviations in profit to understand the underlying factors.
